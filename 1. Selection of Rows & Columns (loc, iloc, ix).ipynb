```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
ufo.info()
```

# loc

```python
#loc is for filtering rows and selecting columns by labels.
#For row we mention index; for column we mention the column names
#A comma seperates the rows and columns you want.
#Let we want 0th row and all columns
ufo.loc[0, :]
```

```python
#Let's say we want from 0th to 2nd row and all columns
ufo.loc[0:2,:]
```

```python
#Another Example
ufo.loc[:,'City']
```

```python
#Another example
ufo.loc[:,['City','State']].head()
```

```python
#If you want columns from 'City' to 'State'
ufo.loc[:, 'City':'State'].head()
```

```python
#Another Example
ufo.loc[0:2,'City':'State']
```

```python
ufo[ufo['City']=='Oakland']
```

```python
#Doing the above with loc
#With loc we define -> we want rows where City = Oakland. And want all columns thus :
#loc is more explicit, thus you can define things as you want
#loc is better for row and column selection and also efficent. So we'll prefer mostly loc method.
ufo.loc[ufo['City']=='Oakland' , :]
```

```python
#Lets say we want rows where City=Oakland and we want columns State and Time
ufo.loc[ufo['City']=='Oakland', ['State','Time']]
```

# iloc

```python
#iloc is for filtering rows and selecting columns by integer positions
#Let's say we want all rows and only 0th and 3rd column
ufo.iloc[:, [0,3]].head()
```

```python
#You can also use range
#Note in loc, range is inclusive on both sides
#But in iloc it is exclusive of the right side thus for 0:4 you get 0,1,2,3 column
ufo.iloc[:,0:4]
```

```python
#Here for 0:3, it will fetch rows 0, 1, and 2
ufo.iloc[0:3, :]
```
